```python
import math
import numpy as np
def evaluacion(x):
  y=1/(x-1)**(1/2)
  return y
h=0.1
b=3
a=2
n=10
count=0
i=1
sumapar=0
sumaimpar=0
print('{:^10}{:^10}{:^10}'.format('i','xi','evaluacion'))
print('{:^10}{:^10}{:^10}'.format('--------------','--------------','--------------'))
print('{:^10}{:^10}{:^10}'.format('0',a,evaluacion(a)))
while i<n:
  xi=a+i*h
  count+=1
  if i%2==0:
    sumapar+=evaluacion(xi)
  else:
    sumaimpar+=evaluacion(xi)
  i+=1
  print('{:^10}{:^10.1f}{:^10.4f}'.format(count,xi,evaluacion(xi)))
resultado=h/3*(evaluacion(a)+evaluacion(b))+(((2*h)/3)*sumapar)+((4*h)/3)*sumaimpar
print('{:^10}{:^10}{:^10}'.format(n,b,evaluacion(b)))
print()
print("la solucion es: ",resultado)
```
